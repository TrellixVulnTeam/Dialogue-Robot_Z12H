```python
# https://blog.csdn.net/Andy_shenzl/article/details/108076056
import numpy  as np
from scipy.stats import norm
import matplotlib.pyplot as plt
X1 = np.linspace(-5, 5, num=20)
gauss1=norm.pdf(X1, loc=0, scale=2) # 均值为0，标准差为2
X2=np.linspace(8, 16, num=20)
gauss2=norm.pdf(X2, loc=12, scale=2) # 均值为12，标准差为2
plt.figure(figsize=(12, 6))
for i in range(len(X1)):
    plt.scatter(X1[i], 0,marker ='x',color='blue')
plt.plot(X1,gauss1)
for i in range(len(X2)):
    plt.scatter(X2[i], 0,marker ='x',color='red')
plt.plot(X2,gauss2)
plt.show()
```

```python
# np random normal https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
n = 10000 # number of sample to be drawn
mu = [0, 12] # 均值
sigma = [2, 2] # 标准差
samples = []
for i in range(n): # iteratively draw samples
    Z = np.random.choice([0,1]) # latent variable
    samples.append(np.random.normal(mu[Z], sigma[Z], 1))
sns.distplot(samples, hist=False)
plt.show()
sns.distplot(samples)
plt.show()
```
